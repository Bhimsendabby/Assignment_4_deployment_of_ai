# Spotify Track Dataset


## These are column name the data contains:

id: Unique identifier for each track.

name: Name of the track.

popularity: Measure of how popular the track is.

duration_ms: Duration of the track in milliseconds.

explicit: Indicates whether the track contains explicit content (0 = not explicit, 1 = explicit).

artists: Name(s) of the artist(s) who created the track.

id_artists: Unique identifier(s) for the artist(s).

release_date: Date when the track was released.

danceability: Measure of how suitable a track is for dancing.

energy: Represents the intensity and activity level of the track.

key: The key the track is in.

loudness: Overall loudness of the track.

mode: Indicates whether the track is in a major or minor key (0 = minor, 1 = major).

speechiness: Detects the presence of spoken words in the track.

acousticness: Likelihood of the track being acoustic.

instrumentalness: Likelihood of the track containing no vocals.

liveness: Indicates the presence of a live audience in the recording.

valence: Describes the musical positiveness conveyed by the track.

tempo: Overall estimated tempo of the track.

time_signature: Number of beats in each bar of the track.

```python
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
columns_to_plot = ['popularity', 'danceability', 'acousticness', 'valence', 'tempo','speechiness','loudness']

for column_name in columns_to_plot:
    plt.figure()  # Create a new figure for each boxplot
    df.boxplot(column=[column_name])
    plt.title(column_name)  # Set the title of the plot to the column name
    plt.show()
```

```python
df.set_index("release_date", inplace=True)
df.index=pd.to_datetime(df.index)
df.head()
```

```python
df[["artists"]].iloc[18]
```

```python
df["duration"]= df["duration_ms"].apply(lambda x: round(x/1000))
df.drop("duration_ms", inplace=True, axis=1)
```

```python
df.duration.head()
```

```python
sample_df = df.sample(int(0.004*len(df)))
```

```python
print(len(sample_df))
```

```python
import seaborn as sns

plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y= "loudness", x = "energy", color = "c").set(title="loudness vs Energy")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y= "popularity", x = "acousticness", color = "b").set(title="popularity vs acousticness")
```

```python
df['dates'] = df.index.get_level_values('release_date')
df['dates'] = pd.to_datetime(df['dates'])
years = df['dates'].dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5, kind="hist").set(title="Noumber of Song per year")
```

```python
total_dr = df.duration
fig_dims = (18, 7)
fig, ax = plt.subplots(figsize=fig_dims)

sns.barplot(x=years, y=total_dr, ax=ax, errwidth=False)
plt.title("Year vs Duration")
plt.xticks(rotation=90)

plt.show()
```

```python
corr_df = df.drop(["key", "mode", "explicit"], axis=1).corr(method="pearson")

plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="black")
heatmap.set_title("Correlation Heatmap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```
